### Data Preprocessing

 __Filling Null Values and Removing Duplicates__

```python
# Calculate percentage of missing values
missing_percentage = df.isna().mean() * 100

# Count of duplicated rows
duplicate_counts = df.duplicated().value_counts()


# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(15, 15))

# Plot percentage of missing values
ax1 = missing_percentage.sort_values(ascending=False).plot(kind='bar', color='red', ax=axes[0])
axes[0].set_ylabel("Percentage Missing")
axes[0].set_title("Pictorial Representation of Missing Values in Data2")

# Add annotations for missing values
for p in ax1.patches:
    ax1.annotate(f'{p.get_height():.2f}%', (p.get_x() + p.get_width() / 2, p.get_height()), 
                 ha='center', va='bottom')

# Plot bar plot for duplicated rows
ax2 = duplicate_counts.plot(kind='bar', color=['skyblue', 'salmon'], ax=axes[1])
axes[1].set_title('Count of Duplicated and Non-Duplicated Rows')
axes[1].set_xticks([0, 1])
axes[1].set_xticklabels(['Non-Duplicated', 'Duplicated'], rotation=0)
axes[1].set_ylabel('Number of Rows')
axes[1].set_xlabel('Row Type')

# Add annotations for duplicated rows
for p in ax2.patches:
    ax2.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2, p.get_height()), 
                 ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
import re
# Function to identify non-integer and NaN values
def filter_non_integers_and_nans(df):
    results = {}
    for column in df.columns:
        def is_non_integer(val):
            if pd.isna(val):
                return True
            val_str = str(val)
            # Check if the value contains any non-numeric characters except a dot or comma
            if re.search(r'[^\d.,]', val_str):
                return True
            return False

        filtered_values = df[column][df[column].apply(is_non_integer)].unique()
        results[column] = filtered_values
    
    return results

# Get filtered results
filtered_results = filter_non_integers_and_nans(df)

# Print filtered results
for column, values in filtered_results.items():
    print(f'{column}: {values}')
    print('-' * 40)
```

```python
# Remove whitespace from column names
df.columns = df.columns.str.strip()

# Remove whitespace from data entries
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

# Convert columns to appropriate data types
numeric_columns = [
    'Lever position', 'Ship speed (v)', 'Gas Turbine (GT) shaft torque (GTT) [kN m]',
    'GT rate of revolutions (GTn) [rpm]', 'Gas Generator rate of revolutions (GGn) [rpm]',
    'Starboard Propeller Torque (Ts) [kN]', 'Port Propeller Torque (Tp) [kN]',
    'Hight Pressure (HP) Turbine exit temperature (T48) [C]',
    'GT Compressor inlet air temperature (T1) [C]', 'GT Compressor outlet air temperature (T2) [C]',
    'HP Turbine exit pressure (P48) [bar]', 'GT Compressor inlet air pressure (P1) [bar]',
    'GT Compressor outlet air pressure (P2) [bar]', 'GT exhaust gas pressure (Pexh) [bar]',
    'Turbine Injecton Control (TIC) [%]', 'Fuel flow (mf) [kg/s]',
    'GT Compressor decay state coefficient', 'GT Turbine decay state coefficient'
]

df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Handle missing values (example: fill with mean, can be adjusted as needed)
df.fillna(df.mean(), inplace=True)
```

### Data Visualization

__1) BoxPlot__

```python
plt.figure(figsize=[19,20])
ax = sns.boxplot(df)
ax.set(xticklabels=[])
plt.show()
```

```python
def remove_outliers_iqr(df, column_names):
    df_clean = df.copy()  

    for column in column_names:
        if pd.api.types.is_numeric_dtype(df_clean[column]):
            Q1 = df_clean[column].quantile(0.25)
            Q3 = df_clean[column].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            df_clean = df_clean[(df_clean[column] >= lower_bound) & (df_clean[column] <= upper_bound)]

    return df_clean
```

```python
columns = [
    'GT Compressor inlet air pressure (P1) [bar]',
    'Fuel flow (mf) [kg/s]', 
    'Turbine Injecton Control (TIC) [%]'
]
df = remove_outliers_iqr(df, columns)
df
```

```python
plt.figure(figsize=[19,20])
ax = sns.boxplot(df)
ax.set(xticklabels=[])
plt.show()
```
